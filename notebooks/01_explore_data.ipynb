### Key Fields

`query`: user entered term in the search box     
`product_id` : Unique id by amazon    
`relevance_label` : Did a human , think was the item relevant for the query   
`product_title` : Title of Item
`product_bullet_point` : Snippet of highlighted item content
`product_brand`: maker of the item
`product_color`: color of product
`query_type`: indicator weather `other` , `parse_pattern`, `behavioral`, `negations`
